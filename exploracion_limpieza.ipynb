```python
def exploracion_csv(archivo):
  
    #Echamos un vistazo
    print("\nPrimeras filas:")
    display(df.head(5))
    print("\n-----------------------------")

    print("\nÚltimas filas:")
    display(df.tail(5))
    print("-----------------------------")

    print("\nFilas aleatorias:")
    display(df.sample(5))
    print("\n-----------------------------")

    print(f"\nEl df tiene {df.shape[0]} filas y {df.shape[1]} columnas")
    print("\n-----------------------------")

    print(f"\nFilas duplicadas: {df.duplicated().sum()}")
    print("\n-----------------------------")

    print(f"\nColumnas: {df.columns}")
    print("\n-----------------------------")

    # nulos y tipos de columnas
    print("\nLos nulos y tipos de columnas que tenemos en el conjunto de datos son:")
    info_columnas = pd.concat([df.isnull().sum(), ((df.isnull().sum()/df.shape[0]) * 100).round(2), df.dtypes], axis=1)
    info_columnas.columns = ['Total Nulos','Porcentaje Nulos', 'Tipo Columna']
    print(info_columnas)
    print("\n-----------------------------")
    
    # valores columnas
    print("\nValores de las columnas: ")
    df_columnas = df.columns
    
    for col in df_columnas:
        unicos = df[col].nunique()
        print(f"\nLa columna {col.upper()} tiene {unicos} valores únicos")
        if unicos <=50:
            print(f"\nLos valores únicos son:")
            print(df[col].unique())
        print(f"\nNúmero de veces que aparece cada valor:")
        print(df[col].value_counts())
        print("\n-----------------------------")
```
